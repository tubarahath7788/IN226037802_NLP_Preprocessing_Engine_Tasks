**Libraries**

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")

print("Model loaded successfully ✅")
```

**Model Loading**

In this project, we used DialoGPT, a conversational AI model developed by Microsoft.

DialoGPT is based on transformer architecture and is trained on large conversational datasets. It is designed to generate human-like responses and maintain conversation context.

```python
# Chat history
chat_history_ids = None

print("Chatbot: Hello! I am your AI assistant. Type 'exit' to end chat.")

while True:
    user_input = input("You: ")

    if user_input.lower() in ['exit', 'quit']:
        print("Chatbot: Goodbye! Have a nice day 😊")
        break

    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

    if chat_history_ids is not None:
        input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1)
    else:
        input_ids = new_input_ids

    chat_history_ids = model.generate(
        input_ids,
        max_length=1000,
        pad_token_id=tokenizer.eos_token_id,
        attention_mask=torch.ones_like(input_ids),
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.75,
        no_repeat_ngram_size=2
    )

    response = tokenizer.decode(
        chat_history_ids[:, input_ids.shape[-1]:][0],
        skip_special_tokens=True
    )

    print("Chatbot:", response)
```

**Chatbot Implementation**

The chatbot is implemented using a loop that continuously takes user input and generates responses using the transformer model.

The conversation history is maintained to provide context-aware responses.

**How it Works**

Tokenizer → converts text into numbers

Model → generates response

Chat History → keeps conversation context

Sampling (top_k, top_p) → makes responses natural
